```python
print('\n[ findall() ]')
p = re.compile('[a-z]+')  # 소문자 문자열 찾기
print(p.findall('I like apple and orange'))
# 결과: 대문자 시작 단어만 제외
```

```python
print('\n[ search() ]')
# : 일치하는 첫 번째 문자열만 리턴
result = p.search('I like apple and orange')
print(result)
print(result.group())   # 일치하는 전체 문자열

result = p.findall('I like apple and orange')
print(result)           # 일치하는 문자열 리스트
```

```python
print('\n[ match 예제1) 전화번호 분석 ]')
# - 여러 번 반복엔 compile !
# - 한 번이 어렵지 만들면 계속 쓰고 좋다
# ---------------------------------------
tel_check = re.compile('^(\d{2,3})-(\d{3,4})-?(\d{4})$')

print(tel_check.match('02-123-1234'))
# => <re.Match object; span=(0, 11), match='02-123-1234'>
print(tel_check.match('053-334-32443')) 
# => None : 맨 뒤 자리수 다름
print(tel_check.match('053950-4554'))   
# => None : - 없음
```

```python
print('\n[ 예제2) .group([idx]) .groups() 메서드 ]')
#  .group(0) : 전체 문자열 (= group() )
#  .group(1) : 첫번째 그룹
#  .groups() : 튜플로 출력
# ---------------------------------------------------
tel_checker = re.compile('^(\d{2,3})-(\d{3,4})-?(\d{4})$')
m = tel_checker.match('02-123-1234')

print(m.groups)     
# <built-in method groups of re.Match object at 0x0000023C7794EAD0>
print(m.group())    # 02-123-1234
print(m.group(0))   # 02-123-1234
print(m.group(1))   # 02
print(m.group(2,3)) # ('123', '1234')
print(m.start())    # 0     : 매칭된 문자열의 시작 인덱스
print(m.end())      # 11    : 매칭된 문자열의 마지막 인덱스+1
```

```python
print('\n[ 예제3) 010으로 시작하기 ]')
# 010, 011, 016 ~ 019으로 시작하는 전화번호 검사
# (?:0|1|[6-9]) :  0, 1, 6-9 중 하나
# ---------------------------------------------
cell_phone = re.compile('^01(?:0|1|[6-9])-(\d{3,4})-(\d{4})$')
# 헷갈리니깐: [1-9]는 숫자 범위, {1,4}는 숫자 개수
print(cell_phone.match('010-123-1234')) # span=(0,12)
print(cell_phone.match('011-123-1234')) # span=(0,12)
print(cell_phone.match('001-123-1234')) # None
print(cell_phone.match('010-1231234'))  # None
```

```python
print('\n[ 전방 탐색(lookahead) ] =============================')
# 전방 긍정 탐색 ( ?=패턴 )
# : 패턴과 일치하면 패턴 앞의 문자열 반환
# (예) 1000원 -> 숫자만 모으기
# 전방 부정 탐색 ( ?!패턴 )
# : 패턴과 일치하지 않으면 패턴 앞의 문자열 반환
# (예) 전화번호 뒷자리 -> {4}(?!-) 해서 마지막, '네' 자리 반환
# -------------------------------------------------------------
lookahead1 = re.search('.+(?=won)', '1000 won')
if (lookahead1 != None):    # 예외 처리
    print(lookahead1.group())   # 1000
else: 
    print('None')

lookahead2 = re.search('.+(?=am)', '2024-02-08 am 09:24:30')
print(lookahead2)   # match='2024-02-08 '

lookahead3 = re.search('\d{4}(?!-)', '010-123-1234')
print(lookahead3)   #  match='1234'
```

```python
print('\n[ 후방 탐색(lookbehind) ] =============================')
# 후방 긍정 탐색 ( ?<=패턴 ): < 가 생김
# : 패턴과 일치하면 패턴 뒤의 문자열 반환
# (예) price: $51 ---- '(?<=:).+' ----> $51
# 후방 부정 탐색 ( ?<!패턴 )
# : 패턴과 일치하지 않으면 패턴 뒤의 문자열 반환
# -------------------------------------------------------
lookbehind1 = re.search('(?<=:).+', 'price: $51')
print(lookbehind1)  # match=' $51'

lookahead2 = re.search('(?<=am).+', '2024-02-08 am 09:24:30')
print(lookahead2)   # match=' am 09:24:30'

lookahead3 = re.search(r'\b(?<!\$)\d+\b', 'I paid $30 for 100 apples')
print(lookahead3)  # match='100'
```

```python
print('\n[ 단일 도메인 내의 이동 ] ==========================')
# [ Wikipedia 페이지 가져오기 ]
# 연관 기사를 검색해 해당 링크에 다시 접속, 반복하는 크롤러 만들기
html = urlopen('https://en.wikipedia.org/wiki/Kevin_Bacon')
bs = BeautifulSoup(html, 'html.parser')

for link in bs.find_all('a'):
    if 'href' in link.attrs:
        print(link.attrs['href'])
# -> 불필요한 내용이 많다
```

```python
# [ 기사 링크의 공통점 ]
#   - div의 'id'='bodyContent' 에 포함
#   - /wiki/ 로 시작
# =>  ^(/wiki/)((?!:).)*$
#   : wiki로 시작하고 : 이 없는 문자열

body_content = bs.find('div', {'id': 'bodyContent'})
# body Content를 포함하는 링크 모아둠

pattern = '^(/wiki/)((?!:).)*$'
for link in body_content.find_all('a', href=re.compile(pattern)):
    # 패턴에 맞는 href를 가진 'a' 태그 중
    if 'href' in link.attrs:    # 왜 href를 다시 확인?
        print(link.attrs['href'])
```

```python
print('\n[ 링크간 무작위 이동 ] ==========================')
#  [ 무작위 링크 선택 ]
import random
pattern = '^(/wiki/)((?!:).)*$'

# random.seed(datetime.datetime.now())
random.seed(None)   # 난수 발생 초기화

def getLinks(articleUrl):
    html = urlopen('https://en.wikipedia.org'+articleUrl)
    bs = BeautifulSoup(html, 'html.parser')
    bodyContent = bs.find('div', {'id': 'bodyContent'})
    wikiUrl = bodyContent.find_all(
        'a', href=re.compile(pattern))
    return wikiUrl

links = getLinks('/wiki/Kevin_Bacon')
print('link 길이: ', len(links))
while  len(links) > 0:
    newArticle =  links[random.randint(0, len(links)-1)].attrs['href']
    print(newArticle)
    links = getLinks(newArticle)
```

```python
# [ 스택과 큐의 차이 ] --------------------------------------
# 1. 큐
#   - 선착순 개념
#   - 가장 먼저 추가된 요소를 가장 먼저 꺼낼 수 있다
# 2. 스택
#   - 후입선출 개념
#   - 가장 최근에 추가된 요소를 가장 먼저 꺼낼 수 있다
#   - 재귀함수는 스택 -> 계속해서 쌓이고 종결이 안됨
# -----------------------------------------------------------
```

```python
print('[ 중복 방지: set(), 재귀함수 ]')
# [set()]
#   - 중복 요소 제거
# [재귀함수]
#   - 함수 내에서 자기 자신을 호출
#   - 파이썬은 1000회로 제한
# -----------------------------------------------------------
pages = set()
count = 0
def getLinks(pageUrl):
    global pages
    global count
    html = urlopen('https://en.wikipedia.org{}'.format(pageUrl))
    bs = BeautifulSoup(html, 'html.parser')
    for link in bs.find_all('a', href=re.compile('^/wiki/')):
        if 'href' in link.attrs:
            if link.attrs['href'] not in pages: # set() 중복 확인
                newPage = link.attrs['href']
                count += 1
                print(f'[{count}]: {newPage}')
                pages.add(newPage)  # set()에 추가
                getLinks(newPage)   # 재귀함수: 메모리 왕창 먹음

getLinks('') 
```

```python
print('\n[ 전체 사이트에서 데이터 수집 ]')
# 반복 접속하면서 특정 데이터 가져오기
# [수집 정보]
#   - 페이지 제목
#   - 첫 번째 문단
#   - 편집 페이지 링크 등
# -----------------------------------------------------------

pages = set()
count = 0
def getLinks(pageUrl):
    global pages
    global count
    html = urlopen('https://en.wikipedia.org{}'.format(pageUrl))
    bs = BeautifulSoup(html, 'html.parser')
    try:    # 예외 처리
        print(bs.h1.get_text())     # 제목 출력
        print(bs.find(id='mw-content-text').find('p').text)
        # 첫 번째 본문 출력
    except AttributeError as e: # : 'mw-content-text' 없을 때
        print('Error:', e)
    
    for link in bs.find_all('a', href=re.compile('^/wiki/')):
        if 'href' in link.attrs:
            if link.attrs['href'] not in pages: # set() 중복 확인
                newPage = link.attrs['href']
                print('-'*40)
                count += 1
                print(f'[{count}]: {newPage}')
                pages.add(newPage)  # set()에 추가
                getLinks(newPage)   # 재귀함수: 메모리 왕창 먹음

getLinks('') 
```
