```python
# 6.1 대소문자 변환과 공백
# 시작하기
import pandas as pd
file1 = '../DATA/chicago_food_inspections.csv'
inspections = pd.read_csv(file1)
inspections
```

```python
inspections['Name'].values  # 공백이 많다
```

```python
# .str.rstrip   .lstrip   .strip  : 좌. 우, 양측 공백 제거
inspections['Name'].str.lstrip().values, inspections['Name'].str.rstrip().values, inspections['Name'].str.strip().values
```

```python
inspections['Name'] = inspections['Name'].str.strip()
inspections['Name'] = inspections['Name'].str.lower()
```

```python
for col in inspections.columns:
    inspections[col] = inspections[col].str.capitalize()
inspections
```

```python
# title() : 각 단어 첫 글자 대문자화
for col in inspections.columns:
    inspections[col] = inspections[col].str.title()
inspections
```

```python
# 6.2 문자열 슬라이싱 : 동일한 자리(문자 인덱스)에 있는 글자를 모아보고 싶다
# 1) 확인
inspections['Risk'].unique()
```

```python
# 2) 결측치 제거
# inspections['Risk'].dropna()  => 이러면 전체 반영할 수 없다~
inspections = inspections.dropna(subset=['Risk'])

# 3) all -> Risk 4
inspections.Risk = inspections.Risk.str.replace('All', 'Risk 4 (Extreme)')
inspections.Risk.unique()
```

```python
# 6.3 문자열 슬라이싱과 문자 치환
# 위에서 정리한 Risk에서 숫자만 가져오자

inspections.Risk.str.slice(5,6)
inspections.Risk.str[5:6]   # same results
```

```python
# slice(start , stop) : 인덱스 start에서부터 stop 까지 출력
inspections.Risk.str.slice(8)
inspections.Risk.str[8:]
```

```python
# 마지막 인덱스 제거
inspections.Risk.str.slice(8, -1)
inspections.Risk.str[8:-1]
inspections.Risk.str.slice(8).str.replace(")","")  # .str 은 매번 써야한다
```

```python
# 6.4 불리언 메서드
"Pizza" in "Mad Pizza"
```

```python
# lower() 로 대소문자 다른 문제를 방지
inspections.Name.str.lower().str.contains("pizza")
```

```python
# pizza 포함 인덱스
inspections[ inspections.Name.str.lower().str.contains("pizza") ]
```

```python
inspections[ inspections.Name.str.lower().str.startswith("tacos") ]
```

```python
inspections[ inspections.Name.str.lower().str.endswith("tacos") ]
```

```python
# 6.5 문자열 분할
customers = pd.read_csv("../DATA/customers.csv")
customers
```

```python
customers.Name.str.len()
```

```python
# split( ) 의 사용
customers.Name.str.split(' ')
customers.Name.str.split(pat=' ')
```

```python
# len() : 리스트의 길이 보기
customers.Name.str.split(' ').str.len()
```

```python
# split(pat= , n =) : n= 분할 개수 제한
customers.Name.str.split(' ', n=1)  # 모두 2개까지만 분할
```

```python
# str.get() : 각 행의 해당 인덱스 원소를 가져옴
customers.Name.str.split(' ', n=1).str.get(0), customers.Name.str.split(' ', n=1).str.get(-1)
# -1 : 마지막 인덱스
```

```python
# expand= : 각 행을 Series 로 뽑아주는 파라미터
# n을 제한하지 않으면 없는 칸에 None 넣음

customers.Name.str.split(' ', n=1, expand=True)
```

```python
# 새 열에 성, 이름 넣기
customers[["First Name", "Last Name"]] = customers.Name.str.split(' ', n=1, expand=True)
```

```python
# 기존 이름 지우기
customers = customers.drop(labels='Name', axis='columns')
customers
```

### 6.6 코딩 첼린지
- 고객 데이터셋에는 주소 열이 있다.
- 각 주소는 도로, 시, 주 및 우편번호로 구성
- 네 가지 값을 분리하고 각 값을 각각 Street, City, State, Zip 이라는 새 열에 할당
- 기존의 Address 열을 삭제

```python
# 앞의 파일을 그대로 사용
customers.Address
```

```python
# 쉼표 구분, 확장
customers.Address.str.split(', ', n=4, expand=True)
```

```python
# 컬럼 지정
customers[["Street", "City", "State", "Zip"]] = customers.Address.str.split(', ', n=4, expand=True)
customers
```

```python
# 기존 주소 삭제
customers.drop(labels='Address', axis='columns')


# del customers["Address"]  => 이거도 가능
```
