## [ 5장 : DataFrame Filtering ]

```python
# 5.1 데이터셋과 메모리 최적화
import pandas as pd
pd.read_csv('../DATA/employees.csv', parse_dates=['Start Date']).head()
```

```python
employees = pd.read_csv(
    '../DATA/employees.csv', parse_dates=['Start Date'])
employees.info()
```

```python
employees['Mgmt'] = employees['Mgmt'].astype(bool) # astype()
employees.tail()
```

```python
employees.info() # 메모리 사용량이 줄었다!
```

```python
employees["Salary"] = employees["Salary"].fillna(0).astype(int)
```

```python
employees.nunique() # 범주형 데이터 확인 -> Gender, Team
employees['Gender'] = employees["Gender"].astype("category")
employees.info()
```

```python
employees['Team'] = employees['Team'].astype("category")
employees.info()
```

```python
# filtering
employees[ employees['First Name'] == "Maria" ]
marias = employees['First Name'] == "Maria" 
employees [marias]
```

```python
employees[ employees['Team'] != "Finance" ]
```

```python
# bool 타입은 바로 필터링이 가능하다
employees[ employees["Mgmt"]].head()
```

```python
high_earners = employees["Salary"] > 10000
employees[ high_earners ].head()
```

```python
# 다중 조건 필터링
# AND
employees[ 
    (employees['Gender']== 'Female') & (employees['Team']=='Business Dev') & employees['Mgmt']
    ].head()
```

```python
# OR
employees[ 
    (employees['Gender']== 'Female') | (employees['Team']=='Business Dev') | employees['Mgmt']
    ].head()
```

```python
# ~
employees[ ~(employees["Salary"] >= 100000)].head()
```

```python
# .isin
employees[ employees['Team'].isin(['Sales', 'Legal', 'Marketing']) ]
```

```python
# .between
employees[ employees['Start Date'].between(
    left= "1980-01-01",
    right= "1990-01-01"
)]
```

```python
# 알파벳도 가능
employees[ employees['First Name'].between("R", "S")]
```

```python
# .isnull() .notnull()
employees['Team'].isnull().head()
# NaT, None 모두 null로 판단
```

```python
# .notnull() : .isnull() 정반대
employees['Team'].notnull().head()
```

```python
# NaN 추출하기
# employees[ employees[col].isnull() ]  => .isnull() 인 행만 추출

no_team = employees['Team'].isnull()
employees[no_team].head()
```

```python
# null 다루기: 다시 파일 불러오기
employees = pd.read_csv('../DATA/employees.csv', parse_dates=['Start Date'])
```

```python
# .dropna() : nan 포함 행 모두 제거
# how='all' : 모두 nan이어야 제거
employees.dropna()
```

```python
# subset= : 특정 열 지정
employees.dropna(subset=['Gender'])
```

```python
# thresh= : nan 값의 개수가 최소 n개 이어야함
employees.dropna(thresh=4)
```

```python
# 5.5 중복 처리

# .duplicated() : 중복값을 식별 / 앞 인덱스와 중복되면 True
# keep= : 유지할 중복값의 위치
print(employees['Team'].head(),
    employees['Team'].duplicated(keep='first').head(),
    employees['Team'].duplicated(keep='last').head(),
    sep='\n\n')

# ~ 를 사용할 수도 있다
```

```python
# .drop_duplicates : 중복값을 제거
# subset, keep 가능
# keep은 사실상 시작을 처음/마지막으로 할지 결정하는 느낌

employees.drop_duplicates(subset=['Team'], keep='last')
```

### 5.6 코딩 첼린지

```python
# file loading
import pandas as pd
file = "../DATA/netflix.csv"
netfDF = pd.read_csv(file)
netfDF
```

##### 1. 데이셋 최적화

```python
# 데이터 체크
netfDF.info()  # 모두 object / date time, cat 로 변경해야함
```

```python
# datetime 변경을 위해 변수 재지정
netfDF = pd.read_csv(file, parse_dates=['date_added'])
netfDF.type.astype('category')
netfDF.info()
```

##### 2. 제목이 'Limitless'인 모든 콘텐츠를 찾으시오

```python
# 조건문
netfDF[ netfDF['title']=='Limitless' ]
```

##### 3. 'Robert Rodriguez' 감족이 제작하고 유형이 'Movie'인 항목

```python
netfDF[ (netfDF['director']=='Robert Rodriguez') & (netfDF['type']=='Movie') ]
```

##### 4. 등록 날짜가 '2019-07-31' 이거나 감독이 'Robert Altman'

```python
netfDF[ (netfDF['date_added']=='2019-07-31') | (netfDF['director']=='Robert Altman') ]
```

##### 5. 감독이 'Orson Welles'. 'Aditya Kripalani' 또는 'Sam Raimi'인 모든 콘텐츠

```python
directors = ['Orson Welles', 'Aditya Kripalani', 'Sam Raimi']
netfDF[ netfDF['director'].isin(directors) ]
```

```python
# 6. 2019년 5월 1일과 2019년 6월 1일 사이에 등록된 모든 콘텐츠
netfDF[ netfDF['date_added'].between('2019-05-01', '2019-06-01', inclusive='both') ]
```

```python
# 7. dictor 열에서 NaN 값이 있는 모든 행을 삭제
netfDF.dropna(subset=['director'])
```

```python
# 8. 콘텐츠를 단 하나만 등록한 날짜를 식별
# netfDF.drop_duplicates(subset=['date_added'])


# 정답 : keep=False 해야한다; 중복된 값을 지워야하기 때문
netfDF.drop_duplicates(subset=['date_added'], keep=False)
```
