## [ 데이터 시각화 맛보기 ]

```python
# 준비하기
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 데이터 준비
x = range(100)
y = [i**2 for i in x]
```

```python
# 1) 그래프 그리기

plt.plot(x,y)
# type : line2D
```



#### <hr>[기본] 선 그래프
- 시간의 변화를 표현
- 연속적/순서 있는 데이터를 시각화할 겨우 사용

```python
# plot(data) => y축 데이터로 사용
# plot(data1, data2) => x축 data1, y축 data2
# 반환값 : 라인/선 그래프 객체 리스트에 담겨서 반환됨
ret = plt.plot( x, y )
type(ret), len(ret), type(ret[0])

# 
plt.show()
```

```python
ret[0].get_xdata()
```

- 선 그래프에서 형식 지정 => str 타입의 포멧 지정
    * '색상형태선스타일' ==> 'ro--' : red circle, dot line
    * r : red, g: green, b: blue, k: black
    * o : circle, ^ : triangle
    * < : left tri. > : right tri. V : down tri.
    * -. : 실선+ 점선

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot:~:text=may%20be%20ambiguous.-,Markers,-character

```python
# 점으로 표현하기 : ro (red o)
x = [i for i in range(1,100,10)]
y = [i**2 for i in x]
plt.plot(x,y, 'ro')
```

```python
# 점선으로 표현
plt.plot(x,y, 'ro--')  # 빨간 원 + 점선
plt.plot(x,y, 'r^--')  # 빨간 삼각형 + 점선
plt.plot(x,y, 'g^--')  # 초록 삼각형 + 점선

# 겹쳐보이네? 여러줄을 다 출력하나 봄
```

```python
# 여러 그래프를 표현하기

x2 = [x//2 for x in x]
y2 = [i**2.1 for i in x]
ret = plt.plot(x, y, 'g^--', x2, y2, 'ro')
```
