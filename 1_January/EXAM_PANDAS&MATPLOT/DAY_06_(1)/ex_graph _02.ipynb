```python
# 준비하기
import matplotlib.pyplot as plt # 그래프 관련 모듈
import matplotlib as mpl        # 패키지 관련
import matplotlib.font_manager as fm # 폰트 관련

import pandas as pd
import numpy as np
```

##### <hr> 아래는 대부분 동일: 다듬기 위주

```python
# 데이터 준비
x = range(100)
y = [i**2 for i in x]
```

- 선 그래프에서 형식 지정 => str 타입의 포멧 지정
    * '색상형태선스타일' ==> 'ro--' : red circle, dot line
    * r : red, g: green, b: blue, k: black
    * o : circle, ^ : triangle
    * < : left tri. > : right tri. V : down tri.
    * -. : 실선+ 점선

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot:~:text=may%20be%20ambiguous.-,Markers,-character

```python
# 점으로 표현하기 : ro (red o)
x = [i for i in range(1,100,10)]
y = [i**2 for i in x]
plt.plot(x,y, 'ro')
```

```python
# 점선으로 표현
plt.plot(x,y, 'ro--')  # 빨간 원 + 점선
plt.plot(x,y, 'r^--')  # 빨간 삼각형 + 점선
plt.plot(x,y, 'g^--')  # 초록 삼각형 + 점선

# 겹쳐보이네? 여러줄을 다 출력하나 봄
```

```python
# 여러 그래프를 표현하기

x2 = [x//2 for x in x]
y2 = [i**2.1 for i in x]
ret = plt.plot(x, y, 'g^--', x2, y2, 'ro')
```

##### <hr> x,y축 레이블 설정 및 눈금 설정

```python
# draw graph
plt.plot(x, y, 'g^--', x2, y2, 'ro')
plt.title('[테스트]')

# ----------------- [ xlabel, ylabel ] ------------------
plt.xlabel('값', loc='right', fontsize= 'large', color='g') 
plt.ylabel('결과', loc='top', fontsize= 'large', rotation = 'horizontal')  
# - loc='top' : y축 상부에 결과 출력
# - color : 색상
# - rotation : 수평, 수직 결정
# +) - mouseover() : 마우스를 올리면 색상이 바뀐다

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot:~:text=available%20Line2D%20properties%3A-,Property,-Description

# [ x,y axis ticks setting ] :눈금 설정
plt.xticks(ticks=[10, 20, 30, 40, 50], labels=['A','B','C','D','E'], color='r')
plt.yticks(ticks=[1000, 2000, 3000, 4000, 5000], labels=['A','B','C','D','E'], color='b')

# - 얘네들 다 *kwargs : 없어도 되는 애들
# - ticks= []  : 축에 표시될 눈금을 표현
# - labels= [] : 눈금의 인덱스(이름)를 설정
# - color=     : 색상변경도 가능
```

##### 범례 설정

```python
# -------------------------- 틀을 짜기 : figure ------------------------------
plt.figure(figsize=(5,5), facecolor='pink')
# - figsize 로 크기 조정
# - facecolor : 배경색

# --------------------------- 그래프 그대로 불러오기 ---------------------------

# draw graph
plt.plot(x, y, 'g^--', x2, y2, 'ro')
plt.title('[테스트]')

# [ xlabel, ylabel ] : x, y axis label
plt.xlabel('값', loc='right', fontsize= 'large', color='g') 
plt.ylabel('결과', loc='top', fontsize= 'large', rotation = 'horizontal')  

# [ x,y axis ticks setting ] :눈금 설정
plt.xticks(ticks=[10, 20, 30, 40, 50], labels=['A','B','C','D','E'], color='r')
plt.yticks(ticks=[1000, 2000, 3000, 4000, 5000], labels=['A','B','C','D','E'], color='b')

# --------------------------- 범례 설정 -------------------------------------

# 범례 설정 : 한쪽 구성에 달기
plt.legend(['green 결과', 'red 결과'], loc='lower right')
# - [label1, label2] : 범례의 라벨 이름
# - loc= : 위치 임의 조정 


# ----------------------- 그래프 안에 텍스트 출력 -----------------------------

# 1) 경계 상자 설정하기 : 셀 색상, 가장자리 색상, 박스 스타일 
box={
    'facecolor':'.75', 
    'edgecolor':'k', 
    'boxstyle':'round', 
    'alpha':0.2
    }

# 2) 텍스트 입력
plt.text(40, 5000, 'Ramdom Graph', backgroundcolor='yellow', bbox= box, verticalalignment='baseline')
# .text( x 좌표, y 좌표, text , *kwargs )
#   - backgroundcolor= : 배경색
#   - verticalalignment : 수직 정렬 (수평선 기준 위아래); 축약어 : ha= 
#   - horizontalalignment : 수평 정렬; 축약어 : va


# ----------------------------- 화살표 설정 -----------------------------------

plt.annotate(
    'Arrow annotate', ha='center', va='bottom', 
    xy=(10, 3000), xytext=(50, 2000),
    arrowprops= {'facecolor':'black', 'shrink':0.05}
)

# annotate : '주석을 달다'는 뜻
# - 아래 내용을 쭉 읽어봄
# (function) def annotate(
#     text: str,
#     xy: Sequence[float],
#     xytext: Sequence[float] = ...,
#     xycoords: str | Artist | Transform | ((...) -> Any) = ...,
#     textcoords: str | Artist | Transform | ((...) -> Any) = ...,
#     arrowprops: dict = ...,
#     annotation_clip: bool | None = ...,
#     **kwargs: Any
# ) -> Annotation

# ------------------------- 격자 추가 -----------------------------

plt.grid(True, lw=2, ls='--', c='.75')

# - lw : 선 굵기
# - ls : 선 모양
# - c = 색상 (투명도)
```
