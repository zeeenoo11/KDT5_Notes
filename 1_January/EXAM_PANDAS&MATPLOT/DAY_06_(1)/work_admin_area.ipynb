### [  JSON 데이터 파일 평탄화 후 저장 및 시각화 ]

```python
# 1. 데이터 준비
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

file = '../DATA/경기도행정구역경계.json'
adminAreaDF = pd.read_json(file)
```

```python
adminAreaNormalizedDF = pd.json_normalize(adminAreaDF.features)  
```

```python
proprtiesNum = adminAreaNormalizedDF['geometry.coordinates'].size

for j in range(proprtiesNum):
    admin = adminAreaNormalizedDF['geometry.coordinates']
    areaCoordinates = admin[j][0]
    coordinatesNum = len(areaCoordinates)
    # 고양시 덕양구는 한 차원이 높으므로
    if coordinatesNum == 1:
        areaCoordinates = admin[j][0][0]
        coordinatesNum= len(areaCoordinates)
    # x, y 축의 리스트 출력
    areaX = [areaCoordinates[i][0] for i in range(coordinatesNum)]
    areaY = [areaCoordinates[i][1] for i in range(coordinatesNum)]

    plt.title(label = adminAreaNormalizedDF["properties.name_eng"][j])
    plt.plot(areaX, areaY)
    plt.show()
```
