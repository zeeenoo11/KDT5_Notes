```python
# 1. start Date 를 인덱스 네임으로 지정
# 2. 데이터 타입을 날짜 타입으로 변경
battles = pd.read_csv(
    "../DATA/revolutionary_war.csv",
    index_col="Start Date",
    parse_dates=["Start Date"]
)
```

```python
### 3.2 Series 정렬
```

```python
# sort_values(by=) : 값을 오름차순 정렬
# by= : 정렬할 열을 보여줌
google.sort_values(by='Close'), pokemon.sort_values(by='Type')

# 일단 문제가 시리즈로 변환이 안되어 교재대로 진행이 잘 안됨
```

```python
# Series 생성 및 정렬 : 알파벳 대문자가 소문자보다 먼저 온다
pd.Series(data= ['Adam', 'adam', 'Ben']).sort_values()
```

```python
# ascending=False : 내림차순 정렬
google.sort_values(by='Close', ascending=False).head()
```

```python
# 알파벳의 내림차순 정리
pokemon.sort_values(by='Type', ascending=False).head()
```

```python
# na_position : NaN 값의 위치를 결정 (first or last)
#   - 기본값 : last
#   - 아래 두 값은 결과 같다
battles.sort_values(by="State")
battles.sort_values(by="State", na_position="last")
```

```python
# "first"
battles.sort_values(by="State", na_position="first")
```

```python
# NaN 제거 : .dropna()
battles.dropna().sort_values(by='State')
```

```python
# .sort_index( [ascending=True] )
pokemon.sort_index()
pokemon.sort_index(ascending=True)
```

```python
battles.sort_index()
```

```python
# 결측값 먼저 표시하고 정렬된 날짜 표시하기
battles.sort_index(na_position="first").head()
```

```python
# 내림차순
battles.sort_index(ascending=False).head()
```

```python
# .nsmallest .nlargest : 가장 작은 값과 가장 큰 값
google.sort_values(by='Close', ascending=False).head()
google.nlargest(columns='Close', n=5)
google.nlargest(columns='Close')  # 셋 다 같은 결과
```
