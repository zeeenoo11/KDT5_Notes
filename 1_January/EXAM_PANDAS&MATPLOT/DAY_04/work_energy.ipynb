- 마지막 열(2017 이후)에 총계 추가

```python
# 1. module loading
import pandas as pd
import numpy as np

# 2. Data setting
file = '../DATA/남북한발전전력량.xlsx'

# 3. read_excel
KEnergyDF = pd.read_excel(file)
KEnergyDF
```

```python
# 1. 북한 열에 신재생 행 추가
KEnergyDF.loc[9] = '-'  # index:9 추가
KEnergyDF['전력량 (억㎾h)'][:5] = '남한'
KEnergyDF['전력량 (억㎾h)'][5:] = '북한'
KEnergyDF['발전 전력별'][9] = '신재생'
KEnergyDF
```

```python
# 2.1 합계 도출 : 두 열을 인덱스화 / 행열 바꾸기
KEnergyDF['합계'] = 0
KEnergyDF_T = KEnergyDF.set_index(["전력량 (억㎾h)", "발전 전력별"]).T
KEnergyDF_T
```

```python
# Test 성공 : sum(axis=1) 이면 행의 합계를 출력한다!
test1 = pd.read_excel(file, index_col=[0,1])
test1['합계'] = test1.replace('-',0).astype('int64').sum(axis=1)
test1
```

```python
# 2.2 바뀐 행열에서 '-' 변경 / 타입 수정 / 합계
KEnergyDF_T.loc['합계'] = KEnergyDF_T.replace('-','0').astype('int64').sum()
KEnergyDF_Final = KEnergyDF_T.T
```
