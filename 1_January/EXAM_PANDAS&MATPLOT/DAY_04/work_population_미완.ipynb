
# [ 시도별 전출입 조사]
- 서울과 제주로 가장 많이 이주한 지역 찾기

```python
# 데이터 준비
import pandas as pd

file = '../DATA/시도별 전출입 인구수.xlsx'

popDF = pd.read_excel(file, index_col=[0,1], skiprows=2, header=None)
popDF
```

```python
popDF['합계'] = 0
popDF
```

```python
# 합계 도출
popDF_T = popDF.T.replace('-',0).astype('int64')
popDF_T.loc['합계'] = popDF_T.sum()
popDF_Final = popDF_T.T
popDF_Final
```

```python
print(popDF_Final.loc['서울특별시'].head())
```

```python
# 서울특별시의 합계를 종합 : 
popDF_Final['합계']
```

<hr> 과제를 잘못이해해서 계산한 총 합계 <hr>

```python
# 총합 계산 ㅠ
# 서울특별시 전입 인원 총합 : 전국, 서울 제외
seoulDict = { }
for loca in popDF_Final.loc['서울특별시'].index:
    if loca != '전국' and loca != '서울특별시':
        seoulDict[loca] = popDF_Final.loc[['서울특별시', loca],'합계']
seoulDict
```

```python
# 총합 계산 ㅠ
# 제주특별자치도 전입 인원 : 전국, 제주 제외
jeju_in_sum = 0
for loc in popDF_Final.loc['제주특별자치도'].index:
    if loc != '전국' and loc != '제주특별자치도':
        jeju_in_sum += popDF_Final.loc['제주특별자치도', loc]['합계']
jeju_in_sum
```
