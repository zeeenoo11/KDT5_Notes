# [DataFrame index 2] : 멀티 인덱스 연습




```python
# 1. 데이터 준비
import pandas as pd

# 2. 데이터 입력
exam = {'이름' :['갓기','현길', '시명', '주영'],
        '국어' : [98, 70, 100, 90],
        '미술' : [80, 20, 40, 80],
        '수학' : [90, 100, 90, 60],
        '체육' : [0, 30, 70, 50]}

# 3. 데이터 저장
examDF = pd.DataFrame(exam)
examDF
```

### ㅇ 행 데이터 추가

```python
# 기존 행 추가 : 갓기 98 80 90 0
examDF.loc[4] = ['갓기', 98, 0, 80, 1]
examDF
```

```python
# 새로운 행 추가
examDF.loc[5] = ['갓기', 98, 67, 85, 3]
examDF
```

### (5-4) 새로운 인덱스 설정 - 방법 4 : .set_index(col1, col2, ...) 멀티인덱스 제어

```python
# 이름, 국어, 수학을 인덱스로 설정
examDF1 = examDF.set_index(['이름', '국어', '수학'])
examDF1
# 똑같은거라 인식하면 출력되지 않음!
```

```python
# 인덱스 속성 확인
examDF1.index
```

### ㅇ 행 데이터 선택 : DF객체변수명.loc[행레이블]/DF객체변수명.iloc[위치인덱스]

```python
examDF1.loc[('갓기', 98, 90)]
```

```python
examDF1.loc[('갓기', 98, 80)]
```

```python
examDF1.loc[('갓기', 98, 85)]
```

```python
# 국어가 98 인 애들만 데려오고 싶다
# examDF1.loc[(98)]  => keyError : 하나만 불러올 순 없음
# 98 을 찾으려면 레벨 설정 필요 => 아래의 .xs
```

```python
# .xs : level= n 에서 key 인 값을 찾는다!
examDF1.xs(98,level=1)
```

### [실습] 두 행을 한 번에 가져오기
- ('갓기',  98,  90), ('주영',  90,  60)

```python
examDF1.loc[[('갓기', 98, 90), ('주영', 90, 60)]]    # 여러 인덱스 호출 -> DataFrame 형태로 나옴
# .loc[ [행인덱스] , [열인덱스] ]
```

### (6) 인덱스 재정렬 : .reindex( ) - 재배치와 추가
- 국어, 미술, 수학, 체육 ... 의 순서를 원하는대로 재배치 + 새로운 행/열 추가
- DF.reindex()


```python
print(examDF.index, examDF.columns, sep='\n\n')
```

```python
# 컬럼별 순서 변경 (얘도 본문을 바꾸진 않음)
examDF.reindex(['체육', '미술', '국어', '수학', '총합'], axis='columns', fill_value=0)
# - axis =       : 기준 행/열
# - fill_value = : 새로운 행/열에 추가할 내용
```

```python
# 행 인덱스 재배치
examDF.reindex([0,1,2,3,4,5,6], axis='index', fill_value=0)
```

```python
# 컬럼별 순서 변경 ==> '체육', '미술', '국어', '수학', '총합' + 행인덱스 => 0,2,4,1,3,5 재배치
examDF.reindex(index=[0,2,4,1,3,5], columns=['체육', '미술', '국어', '수학', '총합'],fill_value=0)
```
