```python
# 1. Module loading
import pandas as pd

# 2. Data prepare
datasD = {'name':['홍 길동','이 나영', '이 윤서', '이 현길' ],
         'age': [10, 41, 2, 22],
         'reg_date': ['2021/01/03', '1992/03/01', '2001/05/08', '2005/09/08']}

datasL = [['홍 길동','이 나영', '이 윤서', '이 현길' ],
         [10, 41, 2, 22],
         ['2021/01/03', '1992/03/01', '2001/05/08', '2005/09/08']]
```

```python
# 3. Save data
# - dict -> DF
df1 = pd.DataFrame(datasD)
df1
```

```python
# - list -> DF
df2 = pd.DataFrame(datasL)
df2
```

```python
# 4. Date info
df2.info()
```

<hr>  5. 데이터 전처리

```python
# 5. 데이터 전처리
# 5-1. 행 <=> 열 치환
df2 = df2.transpose()  # or .T
df2.info() 
```

```python
# 5-2. 실제 데이터와 비교
df2.head()
```

```python
# col 1 : object -> int
df2[1] = df2[1].astype('uint8')  # u-: unsigned(no -) / 8 : 1 byte

# col 2 : -> datetime64[ns]
df2[2] = df2[2].astype('datetime64[ns]')
df2.dtypes
```

```python
# 5-3. 0번 칼럼의 이름을 성과 이름으로 분리
# 1) 0번 칼럼만 추출하기
nameSR = df2[0]
nameSR

# 2) for 문 반복
# for 어쩌구 -> 복잡하다

# 3) 다른 방법 : split( expand=True )
namesDF = nameSR.str.split( expand=True )
```

- .splt( expand=True )하여 늘어난 항을 기존 DF에 입력:
- -> .concat

```python
# 4) 기존 df2에 추가하기
# - pd.concat([s1, s2], key=['s1', 's2'])
# - s1 0 s1[0]
#      1 s1[1]
#   s2 0 s2[0]
#      1 s2[1]   # 이렇게 행으로 합쳐진다

pd.concat([df2, namesDF])  # 원하던 변경이 아님
```

```python
# 4-2) 열로 붙이기 : axis=columns
pd.concat([df2, namesDF], axis='columns', ignore_index=True )

# ignore_index 해야 일련의 인덱스로 나온다
```

```python
# 날짜의 특정 부분만 추출하기
# - .dt : datetime Method

df2[2].dt.year, df2[2].dt.month, df2[2].dt.day
```

```python
pd.concat([df2, df2[2].dt.year], axis=1, ignore_index=True)  # axis=1 : columns
```

```python
# [ join= outer or inner ]
# - 합집합 / 교집합 개념
# - outer : save all cols; insert NaN into empty data
# - inner : only shared cols are remained; NaN no needed
```

```python
# 5-5. 두 개의 DF를 행 방향 데이터 연결
print(pd.concat([df2, namesDF], join='outer'),'\n')  # col=[0,1,2], NaT
print(pd.concat([df2, namesDF], join='inner'),'\n')  # col=[0,1]
```

```python
# ignore_index=True : 행으로 합칠 때 index(row)에도 적용
print(pd.concat([df2, namesDF], join='inner'),'\n')               # idx=[0,1,2,3,0,1,2,3]
print(pd.concat([df2, namesDF], join='inner', ignore_index=True)) # [0,1,2,3,4, ..., 7]
```

```python
# [ .Merge ]
# - 두 개의 DF에서 특정 컬럼을 기준으로 데이터를 합치는 것
# - 두 DF에 합치는 기준이 되는 컬럼을 지정 필요
# - 지정된 컬럼명이 없으면 동일한 컬럼명을 기준으로 합쳐짐
# ---------------------------------------------------

#  .merge(
#         df_left, df_right,
#         how= 'inner',          # 기준 열 양쪽 모두 존재 교집합 (<-> outer)
#         on= None,              # None : 공통 모든 열 키로 설정
#         left_on= 왼쪽 키,
#         right_on= 오른쪽 키 )
```

```python
file1 = '../DATA/stock price.xlsx'
file2 = '../DATA/stock valuation.xlsx'
```

```python
sdf1 = pd.read_excel(file1)
sdf2 = pd.read_excel(file2)
sdf1.info(), sdf2.info()
```

```python
pd.merge(sdf1, sdf2)  
# index : file1=13, file2=10  => result=7
# para- X : merge only b/w 'same ids'
```

```python
pd.merge(sdf1, sdf2, how='outer')

# 없으면 NaN으로 채운다
```

```python
pd.merge(sdf1, sdf2, how='left')

# left : sdf1와 sdf2 교집합 중 'left' : sdf1 차집합은 NaN으로 출력, sdf2 차집합은 사라짐
# right : 정반대 - sdf2의 차집합이 남음
```

```python
pd.merge(sdf1, sdf2, how='cross')

# idx x idx : 각 sdf1 idx이 sdf2 idx2와 병합
# 13 x 10 = 130 idx ;;
#    - 데이터 늘릴 때 사용 -
```
