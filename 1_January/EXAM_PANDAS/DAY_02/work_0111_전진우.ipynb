DataFrame의 컬럼별로 따로 저장하기

```python
columnList = {}
for col in movieDF.columns:
    columnList[col] = movieDF[col]
columnList
```

컬럼별 데이터들의 속성, 즉 index, values, shape, ndim, size, dtypes 정보 출력

```python
for col in movieDF.columns:
    print('---',col,'---')
    print('index :', movieDF[col].index)
    print('values :', movieDF[col].values)
    print('shape :', movieDF[col].shape)
    print('ndim :', movieDF[col].ndim)
    print('size :', movieDF[col].size)    
    print('dtypes :', movieDF[col].dtypes)
    print()
```

컬럼별 인덱스를 변경 실습

```python
for col in movieDF.columns:
    movieDF[col].index = [i for i in range(10, movieDF.shape[0]+10)]
    print('-----', col, '-----')
    print(movieDF[col])
    print()
```

컬럼별 데이터(값) 변경

```python
flag = 0
for col in movieDF.columns:
    if movieDF[col].dtypes == 'int64':
        flag = 1
        for val in range(movieDF[col].size):
            movieDF[col][val+10] += 5
    # elif movieDF[col].dtypes == 'object':
    #     flag = -1
    #     for val in movieDF[col]:
    #         val += 'TTT'
    print(flag)
    print('-----', col, '-----')
    print(movieDF[col])
    print()
```
