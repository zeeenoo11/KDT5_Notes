```python
import pandas as pd
bald_prob = pd.read_csv('Data/bald_probability.csv')
bald_prob.head()
```

```python
bald_prob.info()    # object 엄청 많다
```

```python
bald_prob.corr(numeric_only=True).bald_prob.sort_values(ascending=False) 
# 일단 실수 데이터만 분석
```

```python
# Label Encoder
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
label.fit(bald_prob['gender'])
bald_prob.gender = label.transform(bald_prob['gender'])
bald_prob['gender']
```

```python
# 데이터 타입이 object인 컬럼 추출 -> encoder 화
object_cols = bald_prob.select_dtypes(include=['object']).columns
for col in object_cols:
    label.fit(bald_prob[col])
    bald_prob[col] = label.transform(bald_prob[col])
# bald_prob.info()    # Good

# 함수로 만들기
from sklearn.preprocessing import LabelEncoder
def object_encoder(DF):
    new_DF = DF.copy()
    label = LabelEncoder()
    object_cols = new_DF.select_dtypes(include=['object']).columns
    for col in object_cols:
        label.fit(new_DF[col])
        new_DF[col] = label.transform(new_DF[col])
    return new_DF

# Test function
bald_prob_test = pd.read_csv('Data/bald_probability.csv')
new_bald_DF = object_encoder(bald_prob_test)
new_bald_DF.info()  # good
```

```python
# 원본과 데이터 대조
bald_prob_test.head()
```

```python
new_bald_DF.head()  # 분류가 잘 되어 보인다
```

```python
# 상관 계수
new_bald_DF.corr(numeric_only=True).bald_prob.abs().sort_values(ascending=False).head(7)
# 그래도 상관계수는 비슷하다
```

```python
# Make DataFrame .corr() descending list

def corr_dataframe(dataDF, top5=False):
    data_corr = dataDF.corr().abs()
    df = pd.DataFrame()
    for i in range(len(data_corr.columns)):
        corr_i = data_corr.iloc[:, i].sort_values(ascending=False).reset_index(drop=True)
        df = pd.concat([df, corr_i], axis=1)
    
    if top5:
        mask = df.iloc[1, :].sort_values(ascending=False).head().index # 높은 상관계수를 보이는 col
        return dataDF.corr()[mask]
    else:    
        return df

corr_list = corr_dataframe(new_hair, top5=True)
corr_list
```

```python
# Take Top5 columns
mask = corr_list.iloc[1, :].sort_values(ascending=False).head().index # 높은 상관계수를 보이는 col
new_hair.corr()[mask]
```
