#### [ 파이 Pie plot ]

```python
# 데이터 준비
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 데이터 입력
labels = ['frogs', 'hogs', 'dogs', 'logs']
sizes = [15, 30, 45, 10]

# 1개의 subplot을 ax에 저장하여 반환
fig, ax = plt.subplots()
ax.pie(sizes, labels= labels)

plt.show()
```

```python
plt.subplots()
```

```python
# Auto-label slices
# autopct= : 자동 라벨 입력

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%') # 소수점 한 자리, % 표시
plt.show()
```
