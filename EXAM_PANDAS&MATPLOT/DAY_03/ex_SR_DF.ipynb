# [Series 데이터 연산]

```python
# 1. 모듈 로딩
import pandas as pd
import numpy as np

# 2. 데이터 준비
data = [5, 7, np.nan, 9, 11]

# 3. DataFrame에 저장
sr = pd.Series(data)  # 1차원 데이터
sr.index, sr.shape, sr.ndim, sr.dtype
# dtype: nan으로 인해 float
```

```python
sr.values
```

```python
pd.DataFrame([sr+10, sr-10, sr*5, sr/5, sr>5] ,index=['+','-','*','/','>'])
# columns 넣으면 NaN... 왜지?
# index랑 한 번에 못 넣는 듯
```

```python
(sr > 5) & (sr < 10)

#   &     |     ~     모두 가능
#  and    or   not
```

### (5) 연산 수행 => Series 객체 + Series 객체

```python
data2 = [2,4,6,8,9]

sr2 = pd.Series(data2)
```

```python
print( sr + sr2, sr - sr2, sr * sr2, sr / sr2, sr % sr2, sep='\n\n')
```

```python
print( sr > sr2, sep='\n\n')
```

```python
sr2.index = [2,3,4,5,6]
sr2
```

```python
# 이러면 인덱스가 밀려 올바른 값이 나오지 않는다;
print( sr + sr2, sr - sr2, sr * sr2, sr / sr2, sr % sr2, sep='\n\n')
```

```python
# 해결 방법 : .reset_index
print(sr.reset_index(), sr2.reset_index(drop=True), sep='\n\n')
```

```python
# 5. Series 연산 수행 => 전용 메서드
print( sr.add(5), sr.add(5, fill_value=0), sep='\n\n')
```

```python
print( sr.sub(sr2), sr.sub(sr2, fill_value=0), sep='\n\n')
```

```python
addSR = sr.add(sr2, fill_value=0)
subSR = sr.sub(sr2, fill_value=0)
mulSR = sr.mul(sr2, fill_value=0)
divSR = sr.div(sr2, fill_value=0)
df = pd.DataFrame([addSR, subSR, mulSR, divSR], index=['+','-','*','/'])
df
```

```python
print(df+5, df-5, sep='\n')
```

```python
print(df>5, df<5, sep='\n')
```

- DataFrame과 DataFrame 간 연산 수행 -> 동일 위치의 원소끼리 연산 수행

```python
같은 인덱스에서 연산 진행
df2=df*10
df2
```

```python
print(df+df2, df-df2, df*df2, sep='\n\n')
```

```python
df2.index = ['+','minus','*','div',]
```

```python
print(df+df2, df-df2, df*df2, sep='\n\n')
# 인덱스가 달르면 NaN으로 연산되지 않는다.
```

- 연산 수행 메소드 활용 => 결측치, 즉 NaN 빈칸에 대한 처리를 수행 가능
  - fill_value 파라미터에 값 설정

```python
df.add(df2, fill_value=0)
# 이렇게 메소드로 결측치에 0을 주면:
#     깔끔하게 출력된다.
#     대신 fill_value=0 은 필수
```

```python
df.add(df2)   # fill_values 필수!
```
